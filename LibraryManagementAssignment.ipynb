```python
from re import search
#This a list used for the book entires.

library = [
      {"Title": "The Great Gatsby", "Author": "F. Scott Fitzgerald", "year": 1925, "copies_available": 3},
      {"Title": "To Kill a Mockingbird", "Author":  "Harper Lee", "year": 1960, "copies_available": 5},
      {"Title": "1984", "Author":  "George Orwell", "year": 1949, "copies_available": 2},
      {"Title": "The Catcher in the Rye", "Author": "J.D. Salinger", "year": 1951, "copies_available": 4},
      {"Title": "Pride and Prejudice", "Author": "Jane Austen", "year": 1813, "copies_available": 6},
      {"Title": "The Lord of the Rings", "Author": "J.R.R. Tolkien", "year": 1954, "copies_available": 2}
]

#This is where the book library is and what books avaliable.

borrowed_books = [{"Name": "Wyatt Gaines", "Date": 2024}]

#The lines below are me putting the parameters for the library system of adding, remving books, borrwoing, returning, searching, and exiting the library managment system.
while True:
    print("\nLibrary Management System")
    print("1. Add a Book")
    print("2. Remove a Book")
    print("3. Borrow a Book")
    print("4. Return a Book")
    print("5. Search for a Book by Author")
    print("6. Exit")
    option = input("Enter your option: ")

#These if statments show which option you can chose. The reason for this is to make the output understand why its being chosen.
#In this you see where its adding books and the for loop and if statments are what help display that it is entered and added to the list.

    if option == "1":
      title = input("Title: ")
      author = input("Author: ")
      year = int(input("year: "))
      copies_available = int(input("copies_available: "))

      book_found = False
      for book in library:
        if (book ["Title"] == title and book ["Author"] == author and book ["year"] == year):
          book["copies_available"] += copies_available
          book_found = True
          break

      if not book_found:
        library.append({"Title": title, "Author": author, "year": year, "copies_available": copies_available})
        print(f"Book '{title}' added to the library!")
      print(library)



#This is for removing the book so we have book -= 1 and book ==0 library.remove(book)
#just this small bit a code goes a long way to remove an entore book in the list.

    elif option == "2":
      title = input("Title: ")

      book_found = False
      for book in library:
        if book["Title"] == title:
          library.remove(book)
          book_found = True
          break

      if not book_found:
        print(f"Book '{title}' not found.")
      print(library)

#When you get to option 3 for borrowing a book it has to be able to be seen in when you seach a book that it isn't there so that is why we have the book_found = False.
#The copies_available is the main ticket the book["copies_available"] -= 1 is how the copies available gets subtracted from the actual available books you can see.
#This just shows the code isn't there anymore.

    elif option == "3":
        title = input("Title: ")
        author = input("Author: ")
        year = int(input("year: "))

        book_found = False
        for book in library:
          if (book["Title"].lower() == title.lower() and book["Author"].lower() == author.lower() and book["year"] == year):
            if book["copies_available"] > 0:
              book["copies_available"] -= 1
              borrowed_books.append({"Title": title, "Author": author, "year": year})
              print(f"Book borrowed successfully!")
              print(f"New copies available: {book['copies_available']}")
            else:
              print("No copies of this book are available for borrowing.")
            book_found = True
            break

        if not book_found:
          print(f"Book '{title}' not found in the library.")
        print(library)

#In this one you have the return a book so we again use the book += 1 which just means that you are adding a book that code means it being
#added about it had the book -= 1 menaing the book was being removed.

    elif option == "4":
      title = input("Title: ")
      author = input("Author: ")
      year = int(input("Year: "))

      book_found = False
      for book in library:
        if (book["Title"].lower() == title.lower() and book["Author"].lower() == author.lower() and book["year"] == year):
          book["copies_available"] += 1
          print(f"Book returned successfully!")
          print(f"New copies available: {book['copies_available']}")
          book_found = True
          break

      if not book_found:
        print(f"Book '{title}' not found.")
      print(library)

#In this one the code is saying that the book isn't there anymore or just can be found. So we have the book_found is False. However, when there is a
#book and it is found it will just take you back to the options page.
#I used the search_author so that it can help give the code a more clear direction when finding the book by Author.

    elif option == "5":
        search_author = input("Enter the Author's name: ").strip().lower()

        book_found = False
        for book in library:
          if book["Author"].lower() == search_author:
            print(f"Title: {book['Title']}")
            print(f"Author: {book['Author']}")
            print(f"year: {book['year']}")
            print(f"copies_available: {book['copies_available']}")
            book_found = True
            break

        if not book_found:
          print(f"No books found by author '{search_author.title()}'.")
        print(library)

    elif option == "6":
      print("Exiting the Library Management System.")
      break

    else:
      print("Only chose a number 1 - 6!")

#This is the end this when the book system is finished.
```
